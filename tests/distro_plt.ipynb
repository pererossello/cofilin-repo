```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

$$ 
\gamma = \mu - \sigma^2
$$

```python
mode = 1.

sigma = 0.3
mu = np.log(mode) + sigma**2

# Generate x values
x = np.linspace(0.01, 5, 1000)

# Compute the lognormal PDF
fact1 = (1 / (x * sigma * np.sqrt(2 * np.pi)))
pdf =  fact1 * np.exp(-((np.log(x) - mu) ** 2) / (2 * sigma**2))

# Plot
plt.figure()
plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('LogNormal Distribution (loc=ln(10), scale=0.35)')
plt.show()
```

```python
shape = 2.0
rate = 2.0

# Generate x values
x = np.linspace(0, 10, 1000)

# Compute the Gamma PDF: (rate^shape / Gamma(shape)) * x^(shape-1) * exp(-rate * x)
pdf = (rate**shape) * x**(shape - 1) * np.exp(-rate * x) / math.gamma(shape)

# Plot
plt.figure()
plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('Gamma Distribution (shape=2, rate=2)')
plt.show()
```

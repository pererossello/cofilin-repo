```python
N, L, Z_I, Z_F = 64, 500, 99, 0
cte = Constants(N, L, Z_I, Z_F)
MND = 1e-2
N_TR = L**3 * MND

key = jax.random.PRNGKey(4)
fm_cfg_cwn = FMConfig(
    N,
    L,
    Z_I,
    Z_F,
    lpt_method="1LPT",
    input_kind="CWN",
    cweb=None, 
    det_bias_model='PowerLaw', 
    stoch_bias_model='Poisson',
    N_TR=N_TR
)
fmodel = FModel(fm_cfg_cwn)
q = gen_input_arr(key, cte, fm_cfg_cwn)
din_hat = fmodel.delta_in(q)
din = my_ifft(din_hat, cte.INV_L3)
dev = fmodel.delta_lpt(q)
get_n_tr_mean = fmodel.n_tr_mean()
params = {'alpha': 1.}
params = manage_params(params)
n_tr_mean = get_n_tr_mean(dev, params)
key_n_tr = jax.random.PRNGKey(1)
n_tr_data = fmodel.sample_n_tr(n_tr_mean, key_n_tr, params=params)

din_wf, q_wf = apply_wiener_filter(n_tr_data, cte, return_kind='CWN')

din_wf_2_hat = fmodel.delta_in(q_wf)
din_wf_2 = my_ifft(din_wf_2_hat, cte.INV_L3)
```

```python
plot_cubes(
    [din, n_tr_data, din_wf, din_wf_2],
    cmap=["seismic_r", "gnuplot2", "seismic_r", "seismic_r"],
    vlim=[1e-1, (0, 20), 1e-1, 1e-1],
    idx=N // 2,
    width=1,
    axis=2,
)

compare_pow_spec([din, din_wf, din_wf_2], L, xlog=False)
```
